```python
def return_claims(db_values):
    data = db_values.copy()
    data['key'] = data['channel_group'].str.replace(" ", "", regex=False)
    data['key'] = data['key'] + "_" + data['tier'] + "_"+data['proj_year'].astype(str)
    data.columns = ['proj_year', 'tier', 'channel_group', 'sum(GPO)', 'sum(MAF)', 'sum(claim_counts)', 'key']
    return data
```

```python
def get_db_values():
    
    retail30_proposed = pd.read_csv("retail30_retailSrx_proposed.csv")
    retail90_proposed = pd.read_csv("retail90_proposed.csv")
    mail_mchoice_proposed = pd.read_csv("mail_mchoice_proposed.csv")
    SRX_retail_proposed = pd.read_csv("SRX_retail_proposed.csv")
    speciality_proposed = pd.read_csv("speciality_proposed.csv")

    all_dfs_keys = [retail30_proposed, retail90_proposed, mail_mchoice_proposed, SRX_retail_proposed, speciality_proposed]
    all_dfs_values = ["Retail30", "Retail90", "Mail_Mchoice", "RetailSRX", "Specialty"]
    
    final_db_values = {}
    
    for i in range(len(all_dfs_keys)):
        all_dfs_keys[i]['Tier'] = ['2TQ', '3TNQ', '3TQ', 'Closed', 'Blended']
        result_dict_1 = dict(zip(all_dfs_values[i]+"_"+all_dfs_keys[i]['Tier']+"_1", all_dfs_keys[i]['Unnamed: 2']))
        result_dict_2 = dict(zip(all_dfs_values[i]+"_"+all_dfs_keys[i]['Tier']+"_2", all_dfs_keys[i]['Unnamed: 4']))
        result_dict_3 = dict(zip(all_dfs_values[i]+"_"+all_dfs_keys[i]['Tier']+"_3", all_dfs_keys[i]['Unnamed: 6']))
        
        combined_dict = {**result_dict_1, **result_dict_2, **result_dict_3}
        
        final_db_values = {**final_db_values, **combined_dict}
        
    return final_db_values
```
